0. Load Important modules

```python
import pandas as pd
import numpy as np
from scipy import io
import scipy.special as sc
import os
import read_fixed_input
import Fluspect
import math
import matplotlib.pyplot as plt
```

1. Read Fixed Inputs

```python
fixed = read_fixed_input.read_fixed_input()
spectral = fixed['spectral']
optipar = fixed['optipar']
```

2. Read Leaf Traits

```python
leafbio = io.loadmat("leafbio.mat")
leafbio.keys()
```

3. Run Fluspect with the parameters in "leafbio"

```python
leafopt = Fluspect.Fluspect(spectral,leafbio,optipar)
leafopt.keys()
```

4. Plot the Results

```python
fig, axs = plt.subplots(2, 2,sharex=True,sharey=True)
axs[0, 0].plot(spectral['wlP'], leafopt['refl'])
axs[0, 0].set_title('leafopt refl')
axs[0, 1].plot(spectral['wlP'], leafopt['tran'], 'tab:orange')
axs[0, 1].set_title('leafopt tran')
axs[1, 0].plot(spectral['wlP'], leafopt['kChlrel'], 'tab:green')
axs[1, 0].set_title('leafopt kChlrel')
axs[1, 1].plot(spectral['wlP'], leafopt['kCarrel'], 'tab:red')
axs[1, 1].set_title('leafopt kCarrel')

axs[0, 0].set(ylabel='Response')
axs[1, 0].set(xlabel='Band [nm]', ylabel='Response')
axs[1, 1].set(xlabel='Band [nm]')

fig.tight_layout()
```
